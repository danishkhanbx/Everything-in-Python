Creating Series using:
*   Empty
*   Lists
*   Array
*   Dictionary

```python
import pandas as pd
```

# Empty Series

```python
sd = pd.Series()
```

```python
sd
```

# Series using List

```python
l = [10,20,30,40]
```

```python
sd = pd.Series(l)
```

```python
sd
```

```python
sd = pd.Series(l, index=['a', 'b', 'c', 'd'])
```

```python
sd
```

# Series using Array

```python
import numpy as np
```

```python
a = np.array([1,2,3,4,5])
a
```

```python
sa = pd.Series(a)
sa
```

# Series using Dictionary

```python
d = {'a':10, 'b':20, 'c':30, 'd':40}
pd.Series(d)
```

```python
dt = [('a', 10), ('b', 20), ('c', 30)]
pd.Series(dt)
```
