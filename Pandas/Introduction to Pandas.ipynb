Features of Pandas:
*   High Performance data analysis tool.
*   Working with large data set.
*   Supports or load files with different formats.
*   More flexible to use.
*   Represents in Tabular way(rows and columns).
*   Working on missing data.
*   Indexing - Slicing - Subsetting the large data sets.
*   Merge and join two different datasets easily.
*   Reshape data sets.

```python
import pandas as pd
```

# Series

One Dimensional

Syntax: Series(data, index)

```python
l = [10,20,30,40]
```

```python
pd.Series(l)
```

```python
pd.Series(l, index=['i','ii','iii','iv'])
```

# Data Frame

Two Dimensional

Syntax: DataFrame(data)

```python
d = {'name':['sun','moon','star'], 'size':[100, 10, 1]}
```

```python
pd.DataFrame(d)
```
