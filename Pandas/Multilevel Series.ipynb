# Merging DataFrames with MultiIndexes

```python
close = pd.read_csv('http://bit.ly/smallstocks', usecols=[0, 1, 3], index_col=['Symbol', 'Date']).sort_index()
close
```

```python
volume = pd.read_csv('http://bit.ly/smallstocks', usecols=[0, 2, 3], index_col=['Symbol', 'Date']).sort_index()
volume
```

```python
both = pd.merge(close, volume, left_index=True, right_index=True)
both
```

```python
both.reset_index()
```
