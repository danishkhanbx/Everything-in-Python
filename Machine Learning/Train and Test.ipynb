Separating the data into two groups. One group will be used to train the computer and the other one will be used to test the accuracy rate using KNN model.

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
print(x_train.shape)  # Observations to train and test
```

```python
print(x_test.shape)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
predictions = knn.predict(x_test)
print(predictions)
```

```python
print(y_test)
```

```python
from sklearn import metrics
```

```python
performance = metrics.accuracy_score(y_test, predictions)
print(performance)
```

Finding the best value of k.

```python
k_value = {}
k = 1
while k<=25:
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(x_train,y_train)
  predictions = knn.predict(x_test)
  performance = metrics.accuracy_score(y_test, predictions)
  k_value[k] = round(performance,4)
  k += 1
print(k_value)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(list(k_value.keys()), list(k_value.values()))
plt.xlabel('Values of K')
plt.ylabel('Performance')
plt.show()
```
