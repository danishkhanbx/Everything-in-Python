Statistics using KNN(K nearest neighbors)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
x = iris.data
y = iris.target
```

```python
knn.fit(x,y)
```

```python
print( knn.predict( [ [6.4, 2.8, 5.6, 2.2] ] ) ) 
```
